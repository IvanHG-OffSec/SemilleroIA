**Predicción de éxito de productos completamente nuevos en el mercado**

Cuando una empresa nueva lanza un producto o servicio con un enfoque diferente, es difícil saber si tendrá éxito. Esto se debe a que no hay datos previos que ayuden a predecir si a las personas les va a gustar o si realmente lo van a usar. Además, no se puede anticipar con certeza cómo van a reaccionar los consumidores.

Un ejemplo en el campo de la tecnología podría ser el lanzamiento del primer iPhone. El presentar un teléfono sin teclado físico, con una pantalla completamente táctil y que además combinaba funciones de teléfono, reproductor de música y acceso a internet de una forma más versátil y moderna.

**Matriz de confusion.**

Es una herramienta fundamental para evaluar el rendimiento de modelos de clasificación en aprendizaje automático.

Es una tabla que compara las predicciones del modelo con los valores reales, permitiendo identificar aciertos y errores. Se utiliza tanto para evaluar como para monitorear y gestionar modelos de clasificación.

En una matriz de confusión 2x2, se tienen los siguientes términos clave:

Verdadero Positivo (TP): El modelo predice correctamente la clase positiva.

Verdadero Negativo (TN): El modelo predice correctamente la clase negativa.

Falso Positivo (FP): El modelo predice la clase positiva cuando en realidad es negativa (error tipo I).

Falso Negativo (FN): El modelo predice la clase negativa cuando en realidad es positiva (error tipo II).
